2b. Modify the DataFrame so it only contains data for the US (the native-country column) and for workers with 40 or more hours per week.

```python
data.columns = data.columns.str.lstrip()
data.columns = data.columns.str.replace('-','_') #problem with hyphen in column names/labels, so replace it

data.loc[:,["native_country"]] = data["native_country"].str.strip()   #removing leading space in column values
#alternative: data["native_country"] = data["native_country"].str.strip()  


data = data[data.native_country == 'United-States']   #only data for native_country="United States"
# alternative: data = data[data["native_country"] == "United-States"]

data = data[data.hours_per_week > 39]    #only workers with 40+ hours week

data.index = [i for i in range(len(data))]  #updating the indexes after removing non-"United-States" and hours_per_week < 40 rows

data.head()
```

```python
X = data.drop(columns=["income", "race", "workclass", "hours_per_week"])
y = data.income

X_train, X_test, y_train, y_test = train_test_split(X,y)

classifiers = [
    KNeighborsClassifier(3),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GaussianNB()
    ]

for classifier in classifiers:
    # train the classifier
    classifier.fit(X_train, y_train)  
    # test the trained classifier
    y_output = classifier.predict(X_test)
    # compare the predicted output with the actual output
    print(classifier)
    print(f"score: {f1_score(y_test, y_output, average='weighted'):.3f}")
    print(f"accuracy score: {metrics.accuracy_score(y_test, y_output)}")
    print(metrics.confusion_matrix(y_test, y_output))

df = pd.DataFrame({'Actual': y_test, 'Predicted': y_output})
#df
```
